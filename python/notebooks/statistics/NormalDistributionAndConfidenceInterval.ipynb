```python
%matplotlib inline

import scipy.stats
import numpy as np
import matplotlib.pyplot as plt

# here we define a function that generates random data that has a normal distribution. 
# it then shows a line plot and histogram of the data
def plotNormalAndComputeConfidence(mean, std, num_samples):
    normal=np.random.normal(mean, std, num_samples)
    fig, ax = plt.subplots(1,2)
    fig.set_figwidth(15)
    fig.set_figheight(5)
    ax[0].plot(np.arange(num_samples), normal)
    ax[0].set_title(str(num_samples)+" random samples")
    ax[1].hist(normal,100)
    ax[1].set_title("Histogram of "+str(num_samples)+" samples.")
    print('confidence interval of data')
    print(scipy.stats.norm.interval(0.95, normal.mean(), normal.std()))
    print('confidence interval of mean')
    print(scipy.stats.t.interval(0.95, len(normal)-1, loc=np.mean(normal), scale=scipy.stats.sem(normal)))
    print()
```

```python
mean=60
std=500
numsamples=10

plotNormalAndComputeConfidence(mean, std, 10)
plotNormalAndComputeConfidence(mean, std, 10000)
```
