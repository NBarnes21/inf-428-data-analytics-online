Import matplotlib this is how we always start our python plotting programs

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

The next line of code may look a little weird, however it is important.  '%matplotlib inline' tells python to render the plots below each cell (instead of as a popup), this makes it much easier to keep track of where every plot is drawn

```python
%matplotlib inline
```

Now create some fake data and check the types of the variable we have declared so far.  x and y should be lists and 'plt' should be a module.

```python
x=[1,2,3,4,5]
y=[10, 20, 10, 30, 40]

print(type(x))
print(type(y))
print(type(plt))
```

Now we can draw a simple line plot with one line of code

```python
plt.plot(x,y)
```

It's also easy to add a second line plot, a legend, x and y labels a title and a grid as shown below

```python
y2=[15,21,30,12,88]
plt.plot(x,y,label='one')
plt.plot(x,y2,label='two')
plt.legend()
plt.xlabel('Awesome x')
plt.ylabel('Great y')
plt.title('amazing x vs y')
plt.grid()
plt.show()
```
